```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Sample data: [hours studied]
X = np.array([[1], [2], [3], [4], [5], [6], [7], [8]])
# Labels: 0 = fail, 1 = pass
y = np.array([0, 0, 0, 0, 1, 1, 1, 1])

# Train logistic regression model
model = LogisticRegression()
model.fit(X, y)

# Generate range of study hours for smooth curve
X_test = np.linspace(0, 9, 100).reshape(-1, 1)
y_prob = model.predict_proba(X_test)[:, 1]  # Probability of passing

# Plot data points
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='black', label='Data points (fail=0, pass=1)')

# Plot sigmoid curve (probability of passing)
plt.plot(X_test, y_prob, color='blue', label="Logistic Regression Curve")

# Add decision threshold
plt.axhline(0.5, color='red', linestyle='--', label="Threshold = 0.5")

# Add labels and legend
plt.title("Logistic Regression: Probability of Passing vs Hours Studied")
plt.xlabel("Hours Studied")
plt.ylabel("Probability / Outcome")
plt.legend()
plt.grid(True)
plt.show()
```
